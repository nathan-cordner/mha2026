```python
def get_familysearch_link(url):
    
    # Request page
    page = urlopen(url)
    html_bytes = page.read()
    html = html_bytes.decode("utf-8")    
    soup = BeautifulSoup(html, "html.parser")
    
    # Parse HTML
    try:
        
        header = soup.find_all("div", {"data-testid": "individualSummaryDescription"})[0]        
        # missionary_name = header.find_all("h1")[0].text
        # print(missionary_name)
 
        try: 
            cur_link = header.find(class_="familySearchLink")["href"] 
            return cur_link
        except TypeError:
            print(f"No FamilySearch link for {url}")
            return ""
                            
    except IndexError:
        print(f"Having trouble with {url}")
        return ""

# url = "https://history.churchofjesuschrist.org/chd/individual/aage-lauritz-larsen-1894?lang=eng&timelineTabs=allTabs"
# get_familysearch_link(url)
```

New method -- extract birthplace from JSON

```python
import pandas as pd
import json

from bs4 import BeautifulSoup
from selenium import webdriver

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
def get_birthplace(browser, fs_link):
    try:
        browser.get(fs_link)
    except:
        # InvalidArgumentException -- need to import
        print(f"Issue with finding info for {fs_link}")
        return ""
    try:
        WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'script[type="application/ld+json"]')))
    except TimeoutException:
        try: 
            # wait a little longer?
            WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'script[type="application/ld+json"]')))
        except TimeoutException:     
            print(f"Issue with finding info for {fs_link}")
            return ""
        
    html = browser.page_source
    
    soup = BeautifulSoup(html, "html.parser")
    
    header = soup.find("head")
    
    script_tags = header.find_all("script")
            
    for t in script_tags:
        if "@type" in str(t.string):  
            data = json.loads(t.string)          
            
            if "birthPlace" in data:
                if "address" in data["birthPlace"]:
                    return data["birthPlace"]["address"]
                else:
                    print("missing address field:", data["birthPlace"])
                 
    return ""

def get_missionary_family(browser, fs_link):
    try:
        browser.get(fs_link)
    except:
        # InvalidArgumentException
        print(f"Issue with finding parents for {fs_link}")
        return ["", ""]
        

    try:
        WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "familyContainerCss_f14nfnza")))
    except TimeoutException:
        try: 
            # wait a little longer?
            WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "familyContainerCss_f14nfnza")))
        except TimeoutException:     
            print(f"Issue with finding parents for {fs_link}")
            return ["", ""]
        
    html = browser.page_source

    # look for <div data-testid="parents-family-card" 
    soup = BeautifulSoup(html, "html.parser")
    
    try: 
        parents_card = soup.find_all("div", {"data-testid": "parents-family-card"})[0]
    except IndexError:
        # no parents found
        print(f"No parents found at {fs_link}")
        return ["", ""]
        
        
    # find links to father, mother inside card 
    ul = parents_card.find_all("ul")[0]
    links = ul.find_all("a")
    
    info = []
    
    for link in links:
        # print(link["href"])
        fs_code = link["href"].split("/")[-1] 
        # print(fs_code)
        bio = get_birthplace(browser, f"https://ancestors.familysearch.org/en/{fs_code}")
        info.append(bio)
    if len(info) == 0: # no parents?
        info.append("")
        info.append("")
    if len(info) == 1: # info for one parent?
        info.append("")
    return info

```

```python
# Just birthplace data

df = pd.read_csv("all_clean.csv")
birthplaces = []
browser = webdriver.Chrome()

for i in range(len(df)):
    cur_birthplace = df["birth_place"].iloc[i]
    if pd.isnull(df["birth_place"].iloc[i]):
        if not pd.isnull(df["familysearch_links"].iloc[i]):
            # get birthplace
            cur_birthplace = get_birthplace(browser, df["familysearch_links"].iloc[i])
    birthplaces.append(cur_birthplace)
    if (i + 1 ) % 1000 == 0:
        print("Finished", i + 1)

df["birth_place"] = birthplaces
            
df.info()


```

```python
# loop

from datetime import datetime

def get_fs_data(row_start, row_stop = None):

    df = pd.read_csv("all_clean.csv")

    new_data = {"parent1_birthplace" : [],
                "parent2_birthplace": []}
    
    if row_stop == None:
        row_stop = len(df)

    df = df[row_start:row_stop]

    print("Start time:", datetime.now())

    browser = webdriver.Chrome()

    for i in range(len(df)):
        cur_link = df["familysearch_links"].iloc[i]
        if not cur_link:
            new_data["parent1_birthplace"].append(df["parent1_birthplace"].iloc[i])
            new_data["parent2_birthplace"].append(df["parent2_birthplace"].iloc[i])
        elif not pd.isnull(df["parent1_birthplace"].iloc[i]) or not pd.isnull(df["parent2_birthplace"].iloc[i]):
            # as long as one is not null
            new_data["parent1_birthplace"].append(df["parent1_birthplace"].iloc[i])
            new_data["parent2_birthplace"].append(df["parent2_birthplace"].iloc[i])            
        else:
            info = get_missionary_family(browser, cur_link)

            new_data["parent1_birthplace"].append(info[0])
            new_data["parent2_birthplace"].append(info[1])    

        if (i + 1) % 1000 == 0:
            print(f"processed {i + 1} entries")

    print("Finish time:", datetime.now())

    df["parent1_birthplace"] = new_data["parent1_birthplace"]
    df["parent2_birthplace"] = new_data["parent2_birthplace"]

    df.to_csv("all_clean_updated.csv", index=False)


# get_fs_data(0, 3000)
# get_fs_data(3000, 6000)
# get_fs_data(6000, 9000)

```
